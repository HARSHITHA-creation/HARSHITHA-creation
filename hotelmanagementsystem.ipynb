```python
import matplotlib.pyplot as plt
import mysql.connector
import pandas as pd

# Connecting to MySQL database using mysql-connector-python
connection = mysql.connector.connect(
    user='root',
    password='Harshi@123',
    host='localhost', 
    database='hotelmanagementsystem'
)

# Function to fetch data into a DataFrame
def fetch_data(query):
    cursor = connection.cursor()
    cursor.execute(query)
    rows = cursor.fetchall()
    colnames = [desc[0] for desc in cursor.description]
    df = pd.DataFrame(rows, columns=colnames)
    cursor.close()
    return df

# Query to fetch hotel data for specific cities
hotel_query = """
SELECT city, COUNT(*) AS num_hotels
FROM hotel
WHERE city IN ('City A', 'City B', 'City C', 'City D', 'City E')
GROUP BY city;
"""

# Fetching data and storing it in a DataFrame
hotel_df = fetch_data(hotel_query)

# Close the database connection
connection.close()

# Data Visualization
plt.figure(figsize=(8, 6))
plt.bar(hotel_df['city'], hotel_df['num_hotels'], color='skyblue')
plt.title('Number of Hotels in Selected Cities', fontsize=14)
plt.xlabel('City', fontsize=12)
plt.ylabel('Number of Hotels', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import mysql.connector
import pandas as pd

# Connecting to MySQL database using mysql-connector-python
connection = mysql.connector.connect(
    user='root',
    password='Harshi@123',
    host='localhost', 
    database='hotelmanagementsystem'
)

# Function to fetch data into a DataFrame
def fetch_data(query, params=None):
    cursor = connection.cursor()
    cursor.execute(query, params)
    rows = cursor.fetchall()
    colnames = [desc[0] for desc in cursor.description]
    df = pd.DataFrame(rows, columns=colnames)
    cursor.close()
    return df

# Query to fetch hotel data for selected cities
selected_cities = ['City A', 'City B', 'City C', 'City D', 'City E']
hotel_query = """
SELECT city, COUNT(*) AS num_hotels
FROM hotel
WHERE city IN (%s, %s, %s, %s, %s)
GROUP BY city;
"""

# Fetching data and storing it in a DataFrame
hotel_df = fetch_data(hotel_query, selected_cities)

# Close the database connection
connection.close()

# Data Visualization - Scatter Plot
plt.figure(figsize=(8, 6))
plt.scatter(hotel_df['city'], hotel_df['num_hotels'], color='skyblue')
plt.title('Number of Hotels in Selected Cities', fontsize=14)
plt.xlabel('City', fontsize=12)
plt.ylabel('Number of Hotels', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
In conclusion, a hotel management system plays a crucial role in streamlining various operations within a hotel, from guest bookings to room management and beyond. Here are some key points to consider:

Efficient Operations: A well-designed hotel management system enhances operational efficiency by automating tasks such as reservation management, guest check-ins, housekeeping schedules, and billing processes.

Improved Guest Experience: By providing seamless booking experiences, personalized services, and quick responses to guest requests, a hotel management system contributes to higher guest satisfaction and loyalty.

Enhanced Resource Management: With features for inventory management, staff scheduling, and financial reporting, hotel management systems enable better utilization of resources, leading to cost savings and improved profitability.

Data-Driven Insights: By capturing and analyzing data on guest preferences, booking patterns, and revenue performance, hotel management systems provide valuable insights that can inform strategic decision-making and marketing initiatives.

Scalability and Integration: Modern hotel management systems are scalable and flexible, allowing hotels to adapt to changing business needs and integrate with other systems such as online booking platforms, accounting software, and customer relationship management (CRM) tools.

Streamlined Communication: Centralized communication tools within the system facilitate coordination among hotel staff members, ensuring smooth collaboration across departments.

Compliance and Security: Hotel management systems help maintain compliance with industry regulations and data protection standards, safeguarding guest information and ensuring secure transactions.

Overall, a hotel management system serves as a comprehensive platform that empowers hoteliers to deliver exceptional guest experiences, optimize operations, and drive business growth in a competitive hospitality industry landscape.





```
